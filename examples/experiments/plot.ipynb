```python
import pandas as pd
import seaborn as sns

sns.set()
```

## Actor scalability (exp1)

```python
df_exp1 = pd.read_csv('exp1.csv', sep=';')
df_exp1.shape
```

```python
df_exp1['timeMs'] = df_exp1['timeNs'] / 1000 / 1000
ax = sns.lineplot(data=df_exp1['timeMs'])
ax.set_ylim(0, 5)
```
